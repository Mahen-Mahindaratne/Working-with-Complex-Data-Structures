```python
!pip install hmmlearn
import numpy as np
from hmmlearn import hmm
```

```python
# Coin flip data (Heads = 1, Tails = 0)
observations = [1, 0, 1, 1, 0, 0, 1]
```

```python
# Define possible hidden states (fair/biased)
states = [0, 1]
```

```python
# Emission probabilities (fair: equal heads/tails, biased: more heads)
emission_probs = [[0.5, 0.5], [0.7, 0.3]]
```

```python
# Transition probabilities (stay in same state or switch)
transition_probs = [[0.9, 0.1], [0.2, 0.8]]
```

```python
# Initialize HMM model
model = hmm.MultinomialHMM(n_components=len(states))
```

```python
# Convert observations to NumPy array before reshaping
observations_np = np.array(observations)
```

```python
# Train the model on observations (reshape using NumPy array)
model.fit(observations_np.reshape(-1, 1))
```

```python
# Decode the most likely hidden states sequence
hidden_states = model.predict(observations_np.reshape(-1, 1))
```

```python
# Print the hidden states sequence and interpretation
print("Hidden states:", hidden_states)
print("0: Fair coin, 1: Biased coin")
```

```python
# Print decoded coin states
for i, state in enumerate(hidden_states):
    print(f"Flip {i+1}:", "Fair" if state == 0 else "Biased")
```
